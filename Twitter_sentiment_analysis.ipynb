```python
import re
```

```python
def remove_pattern(pattern,text):
  occur=re.findall("@[\w]*",text)
  for i in occur:
    text=re.sub(i,"",text)
  return text
```

```python
train['new_tweet']=[remove_pattern("@[\w]*",sen) for sen in train['tweet']]
print (train.info())
print(train['new_tweet'])
```

```python
train['new_tweet']=train['new_tweet'].str.replace("[^a-zA-Z#]"," ")
print(train['new_tweet'])
```

```python
train['new_tweet']=train['new_tweet'].apply(lambda x: x.split())
print (train['new_tweet'])
```

```python
train['new_tweet']=train['new_tweet'].apply(lambda x:[w for w in x if w not in stop])
print(train['new_tweet'])
```

```python
from nltk.stem import PorterStemmer
ps =PorterStemmer()
train['new_tweet']=train['new_tweet'].apply(lambda x:[ps.stem(w) for w in x] )
train['new_tweet']=train['new_tweet'].apply(lambda x: ' '.join(x))

print(train['new_tweet'])

np.unique(train['label'],return_counts=True)
```

```python
from sklearn .feature_extraction.text import CountVectorizer
CountVectorizer?
```

```python
vectorizer=CountVectorizer(max_df=0.8,min_df=2,max_features=1000, stop_words='english')
f_data=vectorizer.fit_transform(train['new_tweet'])
pdf_data=pd.DataFrame(f_data.todense())
pdf_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pdf_data,train['label'],test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(solver='lbfgs')
LR.fit(x_train,y_train)

prediction=LR.predict(x_test)
# prediction=LR.predict_proba(x_test)
prediction

from sklearn.metrics import accuracy_score
acc=accuracy_score(prediction,y_test)
print(acc)

from sklearn.metrics import f1_score
acc=f1_score(prediction,y_test)
print(acc)
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
new_pred=sv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(new_pred,y_test)
print(acc)

nn_kernel_accurr=f1_score(new_pred,y_test)
print(nn_kernel_accurr)
```
